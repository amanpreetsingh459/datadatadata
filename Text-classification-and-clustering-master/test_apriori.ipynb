```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from apyori import apriori
```

```python
store_data = pd.read_csv('store_data.csv', header=None)
store_data.info()
```

```python
store_data.head()
```

```python
#create list of list of the dataframe as required by the apriori algorithm
records = []
for i in range(0, 7501):
    records.append([str(store_data.values[i,j]) for j in range(0, 20)])
```

```python
records[2]
```

```python
association_rules = apriori(records, min_support=0.0045, min_confidence=0.2, min_lift=3, min_length=2)
association_results = list(association_rules)
```

```python
print(len(association_results))
```

```python
print(association_results[0])
```

```python
rule_df = pd.DataFrame(columns=['Rule', 'Support', 'Confidence', 'Lift'])
rule_df
```

```python
rule_list = []
support_list = []
confidence_list = []
lift_list = []
```

```python
for item in association_results:

    # first index of the inner list
    # Contains base item and add item
    pair = item[0] 
    items = [x for x in pair]
    print("Rule: " + items[0] + " -> " + items[1])
    rule_list.append(items[0] + " -> " + items[1])

    #second index of the inner list
    print("Support: " + str(item[1]))
    support_list.append(str(item[1]))

    #third index of the list located at 0th
    #of the third index of the inner list

    print("Confidence: " + str(item[2][0][2]))
    confidence_list.append(str(item[2][0][2]))
    print("Lift: " + str(item[2][0][3]))
    lift_list.append(str(item[2][0][3]))
    print("=====================================")
```

```python
rule_df['Rule'] = rule_list
rule_df['Support'] = support_list
rule_df['Confidence'] = confidence_list
rule_df['Lift'] = lift_list
```
