```python
from bs4 import BeautifulSoup
import urllib.request

from selenium import webdriver
```

```python
import traceback
import os
import datetime
```

```python
def get_all_links(url, total_urls, visited_urls):
    #all_links = []
    try:
        resp = urllib.request.urlopen(url)
        visited_urls.append(url)
        print("Visited URL:", url)
        soup = BeautifulSoup(resp, from_encoding=resp.info().get_param('charset'))    
        for link in soup.find_all('a', href=True):
            #print(link['href'])
            total_urls.append(link['href'])
    except:
        print("Error processing URL:", url)
        total_urls.remove(url)
        #print(traceback.format_exc())
        
    if len(list(set(total_urls))) > 1:
        for page in list(set(total_urls)):        
            if page not in visited_urls:
                get_all_links(page, total_urls, visited_urls)

    #else:
    return visited_urls, total_urls
```
